```
import pandas as pd
```

```
df_loc = pd.read_csv('locations.csv')
```

```
df_loc.sample(5)
```

```
dummies = df_loc['genres'].str.get_dummies(sep=',')
dummies
```

```
dummies2 = df_loc['locations'].str.get_dummies(sep=',')
dummies2
```

```
df_loc_genres3 = pd.concat([df_loc, dummies], axis=1)
df_loc_genres3
```

```
df_loc_genres3 = df_loc_genres3.drop(columns=['10 November 2006 (USA)\n', '10 September 1999 (USA)\n', '12 October 2001 (Colombia)\n', '21 November 2018 (USA)\n', '22 October 2020 (Mexico)\n',
                    '5 August 2020 (Mexico)\n', '5 June 2015 (Mexico)\n', 'genres', 'TV Series (2018– )\n', 'TV Series (2020– )\n' ])
df_loc_genres3
```

```
df_loc_genres3 = pd.melt(df_loc_genres3, id_vars=['title','year','rating','locations'], var_name='genre', value_name='one')
df_loc_genres3
```

```
df_loc_genres3['genre'] = pd.Categorical(df_loc_genres3['genre'].apply(str))
```

```
df_loc_genres3.dtypes
```

```
filt_genre = df_loc_genres3['one'] == 1
filt_genre
```

```
df_gen = df_loc_genres3[filt_genre]
df_gen
```

```
# Rating por Generos mas populares del pais
df_gen.groupby('genre')['rating'].median().sort_values(ascending=False)
```

```
movie_rank = df_gen.sort_values('rating', ascending=False)
movie_rank.head(20)
```

```
df_gen.groupby('title')['rating'].count()
```

```
dummies2 = df_gen['locations'].str.get_dummies(sep=',')
```

```
df_gen_loc = pd.concat([df_gen, dummies2], axis=1)
df_gen_loc
```

```
df_gen_loc2 = df_gen_loc.drop(columns=['locations', 'one'])
```

```
df_gen_loc2 = pd.melt(df_gen_loc2, id_vars=['title','year','rating','genre'], var_name='location', value_name='one')
df_gen_loc2
```

```
filt_loc = df_gen_loc2['one'] == 1
```

```
df_final= df_gen_loc2[filt_loc]
df_final
```

```
#Locaciones de peliculas con mejores ratings
loc_rank = df_final.groupby('location')['rating'].median().sort_values(ascending=False)
loc_rank.head(30)
```

```
# Generos con mayor numero de peliculas de la region
df_gen.groupby('genre')['title'].count().sort_values(ascending=False)
```

```
# Locaciones mas utilizadas 
loc_number = df_final['location'].value_counts()
loc_number.head(40)
```

```
relation_gen_loc = df_final.groupby(['genre','location'])[['title']].count()
relation_gen_loc.dropna().sort_values('title', ascending=False).head(30)
```
